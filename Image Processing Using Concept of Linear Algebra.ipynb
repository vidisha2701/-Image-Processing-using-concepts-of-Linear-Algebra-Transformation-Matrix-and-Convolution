```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path="G:\\Dataset\\"
imgpath = path + "download.jpg"
imgpath2 = path + "Image 2.jpg"
# cv2.cvtColor()
img= cv2.imread(imgpath, 1)
img2= cv2.imread(imgpath2, 1)
img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

k= np.array(np.ones((3,3),np.float32))/9 #For Blurring of Image

output=cv2.filter2D(img,-1,k)
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(output)
plt.title('Blurred Image')
plt.show()
```

```python
k= np.array(([0,-1,0],[-1,5,-1],
           [0,-1,0]),np.float32)    #For sharpening of image
output=cv2.filter2D(img,-1,k)
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(output)
plt.title('Sharpened Image')
plt.show()
```

```python
k= np.array(([-1,-1,-1],[-1,10,-1], [-1,-1,-1]),np.float32) #For RB-coloring of the image output=cv2.filter2D(img,-1,k) plt.subplot(1,2,1) plt.imshow(img) plt.title('Original Image') plt.subplot(1,2,2) plt.imshow(output) plt.title('Filtered Image') plt.show()
output=cv2.filter2D(img,-1,k)
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(output)
plt.title('Filtered Image')
plt.show()
```

```python

k= np.array(([0.25,0,-0.25],[0.5,0,-0.5],
           [0.25,0,-0.25]),np.float32)    #For RB-coloring of the image
output=cv2.filter2D(img2,-1,k)
plt.subplot(1,2,1)
plt.imshow(img2)
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(output)
plt.title('Filtered Image')
plt.show()

```
